```python
The Pearson Correlation Coefficient (r) is a measure of the linear relationship between two continuous variables. 
It indicates how strongly and in what direction (positive or negative) the two variables are related.
used between continuous variables
```

```python
import numpy as np

# Sample data
X = np.array([2, 4, 6, 8, 10])
Y = np.array([1, 3, 5, 7, 9])

# Calculate Pearson correlation coefficient
r = np.corrcoef(X, Y)[0, 1]

print(f"Pearson Correlation Coefficient: {r}")
```

```python
from scipy.stats import pearsonr

# Sample data
X = [2, 4, 6, 8, 10]
Y = [1, 3, 5, 7, 9]

# Compute Pearson correlation coefficient and p-value
r, p_value = pearsonr(X, Y)

print(f"Pearson Correlation Coefficient: {r}")
print(f"P-value: {p_value}")
```

```python
Spearman Correlation Coefficient
The Spearman correlation coefficient (œÅ) measures the strength and direction of a monotonic relationship between two variables.
It is useful when the data is not normally distributed or the relationship between variables is not linear but still follows an 
increasing or decreasing trend.
```

```python
import numpy as np
from scipy.stats import spearmanr

# Sample data
X = [10, 20, 30, 40, 50]
Y = [3, 8, 15, 20, 25]

# Compute Spearman correlation coefficient and p-value
rho, p_value = spearmanr(X, Y)

print(f"Spearman Correlation Coefficient: {rho}")
print(f"P-value: {p_value}")
```

```python
Univariate Plots (Single Variable)
Definition: These plots show the distribution of a single variable without considering relationships with other variables.

Common Plots:

Histogram (for numerical data)

Boxplot (for identifying outliers and spread)

Kernel Density Plot (KDE) (for smoothed distribution)

Bar Plot (for categorical data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data
data = np.random.normal(50, 15, 200)  # Normally distributed data

# Histogram
sns.histplot(data, bins=20, kde=True)
plt.title("Univariate Histogram with KDE")
plt.show()
```

```python
Bivariate Plots (Two Variables)
Definition: These plots show the relationship between two variables (X vs. Y).

Common Plots:

Scatter Plot (for numerical relationships)

Line Plot (for trends over time)

Box Plot (for comparing distributions across categories)

Heatmap (for correlation matrices)
```

```python
# Sample data
import numpy as np
X = np.random.rand(100)
Y = 2 * X + np.random.normal(0, 0.1, 100)

# Scatter Plot
sns.scatterplot(x=X, y=Y)
plt.title("Bivariate Scatter Plot")
plt.show()
```

```python
Multivariate Plots (Three or More Variables)
Definition: These plots show relationships between multiple variables simultaneously.

Common Plots:

Pair Plot (scatterplots of all variable combinations)

Heatmap (correlation matrix)

3D Scatter Plot (for three continuous variables)

Violin Plot (combination of boxplot and KDE for multiple groups)
```

```python
# Sample multivariate data
import seaborn as sns
import pandas as pd

# Load sample dataset
df = sns.load_dataset("iris")  

# Pair Plot
sns.pairplot(df, hue="species")  
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate sample data
data = np.random.normal(0, 1, 1000)

# Histogram
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.hist(data, bins=30)
plt.title('Histogram')

# Box plot
plt.subplot(122)
sns.boxplot(data)
plt.title('Box Plot')

plt.tight_layout()
plt.show()
```

```python
# Generate bivariate data
x = np.random.normal(0, 1, 1000)
y = x + np.random.normal(0, 1, 1000)

# Scatter plot
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.scatter(x, y, alpha=0.5)
plt.title('Scatter Plot')

# 2D Kernel Density Plot
plt.subplot(122)
sns.kdeplot(x=x, y=y, cmap="Blues", shade=True)
plt.title('2D Kernel Density Plot')

plt.tight_layout()
plt.show()
```

```python
# Generate multivariate data
data = np.random.randn(100, 4)
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D'])

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Pair plot
sns.pairplot(df)
plt.suptitle('Pair Plot', y=1.02)
plt.show()
```
