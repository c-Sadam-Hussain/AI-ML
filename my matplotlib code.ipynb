```python
# BASIC PLOT

import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.plot(x, y)
plt.show()
```

```python
# STYLE OPTIONS

plt.plot(x, y, color='red', marker='o', linestyle='--')
plt.title("Title")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
```

```python
# MULTIPLE LINES

plt.plot(x, y, label="Line 1")
plt.plot(x, [i**2 for i in x], label="Line 2")
plt.legend()
```

```python
#  BAR CHART

plt.bar(['A', 'B', 'C'], [5, 7, 3])
```

```python
# SCATTER PLOT

plt.scatter(x, y)
```

```python
# BAR CHART

import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=20)

# SAVE PLOT
plt.savefig("myplot.png")
```
