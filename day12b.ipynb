```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

data = {
    'work_hours': [10, 12, 14, 15, 16, 18, 20, 22, 24, 26], 
    'grade': [60, 65, 70, 75, 80, 85, 90, 95, 98, 100] 
}

df = pd.DataFrame(data)

# Calculate Pearson correlation between work_hours and grade
correlation, p_value = pearsonr(df['work_hours'], df['grade'])

# Display the Pearson correlation coefficient
print(f"Pearson Correlation Coefficient: {correlation:.3f}")
print(f"P-value: {p_value:.3f}")

# Create a bivariate plot (scatter plot + regression line)
plt.figure(figsize=(8, 6))

# Plot scatter plot with a regression line
sns.regplot(x='work_hours', y='grade', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})

# Add labels and title
plt.xlabel('Work Hours per Week')
plt.ylabel('Grade')
plt.title('Bivariate Plot: Work Hours vs Grade')

# Display the plot
plt.show()
```
